```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 50)
```

```python
y = np.cumsum(np.random.randn(2, 25))
```

```python
plt.figure(figsize=(12,4))
plt.plot(x, y, 'r-')
plt.show()
```
